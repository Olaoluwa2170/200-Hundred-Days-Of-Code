```python
from random import randint

class Dice():
    def __init__(self, num_side=6):
        self.num_side = num_side
        
    def roll(self):
        return randint(1, self.num_side)
```

```python
import math

vec=[1,4,5,6]
vec2=[1,6,7,6]
vectors=[vec,vec2]
def vector_add(v,w):
    return [v_i+w_i for v_i,w_i in zip(v,w)]

def vector_subtract(v,w):
    return [v_i-w_i for v_i,w_i in zip(v,w)]

def vector_sum(vectors):
    result=vectors[0]
    for vector in vectors[1:]:
        result=vector_add(result, vector)
    return result

def scalar_multiply(c, v):
    return [c*v_i for v_i in v]


def vector_mean(vectors):
     n = len(vectors)
     return scalar_multiply(1/n, vector_sum(vectors))



def dot(v,w):
    return sum(v_i*w_i for v_i,w_i in zip(v, w)) #dot product

def sum_of_square(v):
    return dot(v,v)

def magnitude(v):
    return math.sqrt(sum_of_square(v))

def squared_distance(v, w):
    return dot(vector_subtract(v,w))

def distance(v,w):
    return magnitude(vector_subtract(v,w))
```

```python
def shape(A):
    len_of_rows=len(A)
    len_of_columns=len(A[0]) if A else 0
    return len_of_rows, len_of_columns

def get_rows(A, i):
    return A[i]

def get_colunms(A,j):
    return[A_1[j] for A_i in A]

def mean(x):
    sum(x) / len(x)

def de_mean(x):
    x_bar=mean(x)
    return [x_i-x_bar for x_i in x]
def variance(x):
    deviation = de_mean(x)
    n=len(x)
    return[sum_of_square(deviation)/(n-1)]

def standard_deviation(x):
    math.sqrt(variance(x))
    
def covariance(x,y):
    n=len(x)
    return dot(de_mean(x), de_mean(y)/(n-1))

"""dot sums up the products of corresponding pairs of elements."""

def correlation(x, y):
     stdev_x = standard_deviation(x)
     stdev_y = standard_deviation(y)
     if stdev_x > 0 and stdev_y > 0:
         return covariance(x, y) / stdev_x / stdev_y
     else:
         return 0
```
